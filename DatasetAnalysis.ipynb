```python
import matplotlib.pylab as plt
from matplotlib.axes import Axes
import itertools
%matplotlib inline

x, y = list(cnt2.keys()), list(cnt2.values())
x, y = zip(*sorted(zip(x, y)))
plt.plot(x, y, linestyle='-', color='k', marker='o', markerfacecolor='c', linewidth=4, markersize=7)
#plt.bar(x, y, width=1, color='purple', edgecolor='black')
plt.xlabel('Количество изображений в классе', size=15)
plt.ylabel('Число классов', size=15)
plt.title('Все классы, кроме new whale', size=15)

plt.savefig('all_classes.png', bbox_inches='tight')
```
